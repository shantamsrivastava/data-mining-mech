## Importing Libraries

```python
import pandas as pd
import numpy as np
import copy
```

## Data Handling

```python
data = pd.read_excel('Historical Product Demand_project 5.xlsx')
data.head()
```

## Removing Rows with empty values

```python
def data_fix(dataf1):
    dataf = copy.copy(dataf1)
    dataf = dataf.dropna(inplace = False)
    return dataf
```

## Splitting the Data into Training and Testing Data

```python
#Inputs the Data After Feature Engineering Tasks and Splits it into Training and Testing Data, Function Returns two DataFrames
def split(dataf):
    #dataf = dataf.sample(frac = 1) #Shuffles the DataFrame Along the Rows
    ratio = 0.67
    rows = dataf.shape[0]
    train_size = int(rows * ratio)
    train = dataf[0:train_size]
    test = dataf[train_size:]
    return train,test
```

## Final Dataset

```python
def final_data(dataf1):
    dataf = copy.copy(dataf1)
    dataf = data_fix(dataf)
    dataf = dataf.sort_values('Date')
    train, test = split(dataf)
    return train, test
```

```python
train, test = final_data(data)
```
